```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Setze den relativen Pfad zur lokalen CSV-Datei
file_path = "sales_data.csv"

print(f"Lade Datensatz aus lokaler Datei: {file_path}")

try:
    # Lade die CSV in einen Pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Zeige die ersten 5 Zeilen an
    print("Erste 5 Zeilen:")
    display(df.head())
    
    # Zeige grundlegende Informationen zum Datensatz an
    print("\nInformationen zum Datensatz:")
    df.info()
    
except FileNotFoundError:
    print(f"Fehler: Die Datei '{file_path}' wurde nicht gefunden.")
    print(f"Aktuelles Verzeichnis: {os.getcwd()}")
    print(f"Dateien im Verzeichnis: {os.listdir('.')}")
except Exception as e:
    print(f"Ein Fehler ist aufgetreten: {e}")
```

```python
# 1. Datensatz nur auf Deutschland filtern
df_germany = df[df['Country'] == 'Germany'].copy()
print(f"Ursprüngliche Datensätze gesamt: {len(df)}")
print(f"Datensätze für Deutschland: {len(df_germany)}")
```

```python
# 2. Datums-Spalte in das richtige Format umwandeln
df_germany['Date'] = pd.to_datetime(df_germany['Date'])
```

```python
# 3. Date-Shifting (Der Profi-Trick für aktuelle Daten)
# Finde das höchste Jahr im alten Datensatz heraus
max_year_in_data = df_germany['Date'].dt.year.max()
ziel_jahr = 2026 # Das Jahr deiner Simulation
jahre_differenz = ziel_jahr - max_year_in_data

print(f"\nAltes maximales Jahr: {max_year_in_data}")
print(f"Verschiebe alle Daten um {jahre_differenz} Jahre in die Zukunft...")
```
